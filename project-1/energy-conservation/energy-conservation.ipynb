# Energy conservation for varying timesteps

In this notebook we explore how the energy conservation of a system of Argon gas changes as we vary the timestep $\Delta t$. In other words, we are testing the stability of the simulation.

```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)
```

Below we plot energy as a function of time $E(t)$ for $\Delta t = \{0.01, 0.001, 0.0001, 0.00001\}$.

```python
fig = plt.figure(figsize=(14, 10))

for i in range(2, 6):
    dt = 10 ** (-i)
    filename = os.path.join(f"thermo_{i}.log")
    log_df = pd.read_csv(filename, sep="\s+", skiprows=16, skipfooter=28, engine="python")

    plt.plot(log_df.Step * dt, log_df.TotEng, label=fr"$\Delta t = {dt}$")

plt.title(r"Energy as function of time $E(t)$ for different $\Delta t$")
plt.xlabel(r"$t$")
plt.ylabel(r"$E(t)$")
plt.legend(loc="best")
plt.show()
```
